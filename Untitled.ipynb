# 1. Message extraction

### The two snippets that follow have been extracted from Aman Kharwal's post on The Clever Programer, the link of which can be found here https://thecleverprogrammer.com/2021/04/09/whatsapp-chat-analysis-with-python/

```python
def date_time(s):
    pattern = '^([0-9]+)(\/)([0-9]+)(\/)([0-9]+), ([0-9]+):([0-9]+)[ ]?(AM|PM|am|pm)? -'
    result = regex.match(pattern, s)
    if result:
        return True
    return False

def find_author(s):
    s = s.split(":")
    if len(s)==2:
        return True
    else:
        return False

def getDatapoint(line):
    splitline = line.split(' - ')
    dateTime = splitline[0]
    date, time = dateTime.split(", ")
    message = " ".join(splitline[1:])
    if find_author(message):
        splitmessage = message.split(": ")
        author = splitmessage[0]
        message = " ".join(splitmessage[1:])
    else:
        author= None
    return date, time, author, message
```

```python
data = []
conversation = 'adolf-data.txt'
with open(conversation, encoding="utf-8") as fp:
    fp.readline()
    messageBuffer = []
    date, time, author = None, None, None
    while True:
        line = fp.readline()
        if not line:
            break
        line = line.strip()
        if date_time(line):
            if len(messageBuffer) > 0:
                data.append([date, time, author, ' '.join(messageBuffer)])
            messageBuffer.clear()
            date, time, author, message = getDatapoint(line)
            messageBuffer.append(message)
        else:
            messageBuffer.append(line)
```

```python
df = pd.DataFrame(data, columns=["date", "time", "author", "message"])
```

```python
df.date = pd.to_datetime(df.date)
```

#### There a few @s that are not of interest, and as can be seen, they are those who have less than 3 mentions.

```python
mentions = mentions[mentions.groupby('mention').mention.transform('count')>3]
```

### Lets see who has mentioned who - To make it more dynamic I have created a function to return a graph of the author we want to check 

```python
def who_mentions_who(name):
    '''
    This function returns a graph with the people mentioned by a selected author.
    Arg: Name of the author we want to check
    '''
    return mentions.mention[mentions["author"]==name].value_counts().plot(kind="barh")
```
